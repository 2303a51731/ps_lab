```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
N=1000
```

```python
np.random.seed(0)
vechicle_counts = np.random.randint(0,6,N)
```

```python
values,counts = np.unique(vechicle_counts,return_counts=True)
pmf=counts/N
```

```python
cdf=np.cumsum(pmf)
```

```python
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
sns.barplot(x=values,y=pmf,palette="viridis")
plt.title('Probability Mass Function(PMF)')
plt.xlabel('Number of Vehicles')
plt.ylabel('Probability')
```

```python
plt.figure(figsize=(12,6))
plt.subplot(1,2,2)
sns.lineplot(x=values,y=cdf,color='red')
plt.title('Probability Mass Function(PMF)')
plt.xlabel('Number of Vehicles')
plt.ylabel('Cumulative Probability')
```

```python
plt.subplot(1,2,2)
sns.lineplot(x=values,y=cdf,marker="o",color='b')
plt.title('Cumulative Distribution Function(CDF)')
plt.xlabel("Number of Vehicles")
plt.ylabel('Cumulative Probability')
plt.xticks
plt.tight_layout()
plt.show()
```
