```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def load_data(file_path):
  df=pd.read_csv(file_path)
  return df
```

```python
def plot_histogram(df, column, bins=10):
  plt.figure(figsize=(8,6))
  plt.hist(df[column], bins=bins)
  plt.title(f'Histogram of {column}')
  plt.xlabel(column)
  plt.ylabel('Frequency')
  plt.show()
```

```python
def plot_scatter(df,x_column,y_column):
  plt.figure(figsize=(8,6))
  plt.scatter(x=df[x_column],y=df[y_column],color='red')
  plt.title(f'Scatter Plot of {x_column} vs {y_column}')
  plt.xlabel(x_column)
  plt.ylabel(y_column)
  plt.show()
```

```python
def plot_bar_chart(df, x_column, y_column): # Changed y_colomn to y_column
    plt.figure(figsize=(10, 6))
    sns.barplot(x=df[x_column], y=df[y_column], palette='viridis') # Now using y_column
    plt.title(f'Bar Chart of {x_column} vs {y_column}')
    plt.xlabel(x_column)
    plt.ylabel(y_column)
    plt.xticks(rotation=45)
    plt.show()
```

```python
file_path='cricket_dataset.csv'
data=load_data(file_path)
```

```python
plot_histogram(data,'Runs')
plot_histogram(data,'Wickets')
```

```python
plot_scatter(data, 'Matches', 'Runs')
plot_scatter(data, 'Economy Rate', 'Wickets')
```

```python
plot_bar_chart(data, 'Player', '50s')
plot_bar_chart(data, 'Player', '100s')
```
